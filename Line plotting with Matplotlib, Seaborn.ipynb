**Importing Libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Reading CSV File**

```python
df = pd.read_csv(r'J:\Data Science\Projects\Plotting\food.csv')
df.head(10)
```

 **Droppimg unnecessary Columns**

```python
df = df.drop(['ingredients','diet','ingredients','prep_time','flavor_profile','course','state','region'],axis = 1)
df
```

**Creating a dataframe**

```python
food = df.head(15)
food
```

```python
xlabel = food['name']
ylabel= food['cook_time']
```

# LINE Plotting

# Matplotlib

```python
fig = plt.figure(figsize=(25,10))
plt.plot(xlabel, ylabel)
```

```python
fig = plt.figure(figsize=(22,15))
plt.style.use('classic') 
plt.plot(xlabel,ylabel, lw=6, color='green', marker='o', markersize=15, linestyle='solid')
#labeling the axes and set title
plt.title('Food coooking duration')
plt.xlabel('Food Name')
plt.ylabel('Cooking time')
plt.show()
```
